```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import time
```

```python
# create a df of 3000*3000 random numbers
df = pd.DataFrame(np.random.randint(0,100,size=(3000, 3000)), columns=list(range(3000)))
ram_lst = []
```

```python
# process memory size
import os
os.getpid()
import psutil
process = psutil.Process(os.getpid())
# with  comma separator for 1000
ram = f'{process.memory_info().rss:,}'
ram_lst.append(ram)
```

each time you run the next cell, the graph will be updated and you can see the process memory use keeps going up. This is the main problem that concerns me.

```python
with open('test.pkl', 'wb') as f:
    pickle.dump(df, f)
ram = f'{process.memory_info().rss:,}'
ram_lst.append(ram)


with open('test.pkl', 'rb') as f:
    df = pickle.load(f)
ram = f'{process.memory_info().rss:,}'
ram_lst.append(ram)


plt.plot(ram_lst)
plt.show()
```
